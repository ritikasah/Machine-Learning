```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder #charcyers like Y/N are used to enode inot 0 or 1 to make machine understand
from sklearn.model_selection import train_test_split #to train test data 
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score, confusion_matrix #to calculate accuracy of the model as well as other metrics
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('mushrooms.csv')
df
```

as we can see:
we are unable to read the data on our own but the class here describes e=edible and p=poisonous
in order to make the machine understand the data , we need to convert it into numerical data

```python
#to convert categorical data into numerical value
label=LabelEncoder()
for i in df.columns:
    df[i]=label.fit_transform(df[i])
df
```

```python
#splitting into train test
x=df.drop(['class'],axis=1)
y=df[['class']]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=21) #to split into 70-30% using test size as 0.3
LogModel=LogisticRegression(max_iter=500)
LogModel.fit(x_train,y_train)
pred=LogModel.predict(x_test)
```
