```python
import pandas as pd #dataframe which helps in viewing the data in to tabular format
from sklearn.preprocessing import LabelEncoder #charcyers like Y/N are used to enode inot 0 or 1 to make machine understand
from sklearn.model_selection import train_test_split #to train test data 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import accuracy_score, confusion_matrix #to calculate accuracy of the model as well as other metrics
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv("insurance[1].csv")
df
```

As you can see:
1. dimension of the dataset is 1338 rows and 7 columns
2. the attribute sex and smoker are categorical values either f/m or y/n hence label encoder will be used to convert into binary values 
3. region also has the same concept but it will not convert just into 0 and 1 because it has four regions hence will be turned into 0-3

```python
#preprocessing the data
lb=LabelEncoder()
#for sex
lb.fit(df.sex.drop_duplicates()) #to only pick unique values and convert them
df.sex=lb.transform(df.sex)
#for smoker
lb.fit(df.smoker.drop_duplicates())
df.smoker=lb.transform(df.smoker)
#for region
lb.fit(df.region.drop_duplicates())
df.region=lb.transform(df.region)
df
```

```python
#splitting into train test
x=df.drop(['charges'],axis=1)
y=df[['charges']]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=23) #to split into 70-30% using test size as 0.3
LinearModel=LinearRegression()
LinearModel.fit(x_train,y_train)
pred=LinearModel.predict(x_test)
```

```python
#to check the coefficient that the model generates and fits to satisfies the eqn
for idx,col_name in enumerate(x_train.columns):
    print("The coeff for {} is {}".format(col_name,LinearModel.coef_[0][idx]))
```

As we can see:
1. The impact of being smoker impacts the model hugely
2. the bmi also effects to an extent
